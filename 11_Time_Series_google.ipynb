# Resampling

1. Downsampling

Weekly

```python
google['Close'].plot()
google['Close'].resample('W').mean().plot(legend=True)
```

Quaterly

```python
google['Close'].plot()
google['Close'].resample('Q').mean().plot(legend=True)
```

Monthly

```python
google['Close'].plot()
google['Close'].resample('M').mean().plot(legend=True)
```

```python
google['Close'].plot()
google['Close'].resample('Y').mean().plot(legend=True)
```

2. Upsampling

```python
# google['Close'].plot()
google['Close'].resample('h').interpolate(method='spline',order=2).plot()
```

# Rolling (Smoothing)

Moving Average

```python
google['Close'].plot(legend=True)
google['Close'].rolling(60).mean().plot(legend=True)
```

Exponential Moving Average

```python
# google['Close'].plot()
google['Close'].rolling(60).mean().plot()  #blue
google['Close'].ewm(60).mean().plot()    #orange
```
