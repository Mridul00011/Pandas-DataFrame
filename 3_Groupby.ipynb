1. find no of movies starting with A for each group

```python
def start(group):
    return (group['Series_Title'].str.startswith('A')).sum()
```

```python
genres.apply(start)
```

2. Find Ranking of each movie in the group according to TMDB_Rating

```python
def rank(group):
    group['Genre_Rank']=group['IMDB_Rating'].rank(ascending=False)
    return group
```

```python
genres.apply(rank)
```

3. Find Normalised IMDB Rating Group



```python
def normal(group):
    group['norm_rating']=(group['IMDB_Rating']-group['IMDB_Rating'].min()) / (group['IMDB_Rating'].max()-
                                                                              group['IMDB_Rating'].min())
    return group
```

```python
genres.apply(normal)
```

1. Find the top 10 batsman in terms of runs

```python
ipl.head(2)
```

```python
ipl.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).reset_index()
```

2. Find the batsman with max no of sizes

```python
six=ipl[ipl['batsman_runs']==6]
six.head(2)
```

```python
six.groupby('batsman')['batsman_runs'].count().sort_values(ascending=False)
```

3. Find batsman with most number of 4's an 6's in last 5 overs

```python
over=ipl[(ipl['over']>15) & ((ipl['batsman_runs']==4) | (ipl['batsman_runs']==6))]
over.groupby('batsman')['batsman'].count().sort_values(ascending=False)
```

4. Find V Kohli's record against all teams

```python
ipl.info()
```

```python
vk=ipl[ipl['batsman']=='V Kohli']
vk.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False).reset_index()
```

5. Create a function that can return the highest score of any batsman

```python
def highest_score(batsman):
    player=ipl[ipl['batsman']==batsman]
    return (player.groupby('match_id')['batsman_runs'].sum().sort_values(ascending=False))
```

```python
highest_score('S Dhawan')
```
