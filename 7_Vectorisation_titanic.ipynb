# Working On Dataset

```python
df=pd.read_csv('titanic.csv')
df['Name']
```

# Functions in Vectorisation

1. lower()

```python
df['Name'].str.lower()
```

2. upper()

```python
df['Name'].str.upper()
```

3. capitalize()

```python
df['Name'].str.capitalize()
```

4. title()

```python
df['Name'].str.title()
```

5. len()

```python
df['Name'].str.len().max()
```

```python
df['Name'][df['Name'].str.len()==82].values[0]
```

6. strip()

```python
df['Name'].str.strip()
```

7. split()

```python
df['lastname']=df['Name'].str.split(',').str.get(0)
```

```python
df.head()
```

Imp to note that after how many words you want to split

```python
df['Name'].str.split(',').str.get(1).str.strip().str.split(' ',n=1)
```

Imp to note that convert above data in dataframe using parameter expand in split

```python
df[['title','firstname']]=df['Name'].str.split(',').str.get(1).str.strip().str.split(' ',n=1,expand=True)
```

```python
df['title'].value_counts()
```

8. replace()

```python
df['title']=df['title'].str.replace('Mlle.','Miss.').str.replace('Ms.','Miss.').value_counts()
```

```python
df.head()
```

Search John in both cases

```python
df[df['firstname'].str.contains('john',case=False)].head()
```

Find last names with starts and ends with vowel

```python
df[df['firstname'].str.contains('^[aeiouAEIOU].+[aeiouAEIOU]$')]
```
