# date_range()

```python
pd.date_range(start='2024-9-5',end='2024-10-5',freq='3D')
```

B--> Print Business Days

```python
pd.date_range(start='2024-9-5',end='2024-10-5',freq='B')
```

W --> One week per day

```python
pd.date_range(start='2024-9-5',end='2024-10-5',freq='W')
```

```python
pd.date_range(start='2024-9-5',end='2024-10-5',freq='W-THU')
```

H--> Hourly data

```python
pd.date_range(start='2024-9-5',end='2024-10-5',freq='h')
```

ME--> Month End

```python
pd.date_range(start='2024-9-5',end='2024-10-5',freq='ME')
```

MS--> Month Start

```python
pd.date_range(start='2024-9-5',end='2024-10-5',freq='MS')
```

YE--> Year End

```python
pd.date_range(start='2024-9-5',end='2024-10-5',freq='YE')
```

YS--> Year Start

```python
pd.date_range(start='2024-9-5',end='2024-10-5',freq='YS')
```

using period (no. of results)

```python
pd.date_range(start='2024-1-11',periods=25,freq='YS')
```
